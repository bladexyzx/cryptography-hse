```python
def mod_inverse_3x3(matrix, modulus=26):
    """Находит обратную матрицу 3x3 по модулю."""
    M = np.array(matrix, dtype=int)
    
    det = int(np.round(np.linalg.det(M))) % modulus
    
    det_inv = None
    for i in range(modulus):
        if (det * i) % modulus == 1:
            det_inv = i
            break
            
    if det_inv is None:
        raise ValueError(f"Детерминант {det} не имеет обратного элемента по модулю {modulus}.")
        

    adj = np.round(np.linalg.inv(M) * np.linalg.det(M)).astype(int)
    
    inv_matrix = (det_inv * adj) % modulus
    return inv_matrix

```
